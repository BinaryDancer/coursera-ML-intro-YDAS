```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wine.data', header=None)
```

```python
df
```

```python
y = df[0]
X = df.drop([0], axis=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.utils import shuffle
from sklearn.preprocessing import scale
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
# X, y = shuffle(X, y, random_state=42)
result = []
for n in range (1, 50):
    clf = KNeighborsClassifier(n_neighbors=n)
    scores = cross_val_score(clf, X, y, cv=kf)
    result.append(scores.mean())
    print(n, scores, scores.mean())
print('max score without *scale* : score = {}, N = {}'.format(
    result[np.argmax(result)], 
    np.argmax(result) + 1))
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
X1 = scale(X)
result1 = []
for n in range (1, 50):
    clf = KNeighborsClassifier(n_neighbors=n)
    scores = cross_val_score(clf, X1, y, cv=kf)
    result1.append(scores.mean())
    print(n, scores, scores.mean())
print('max score with *scale* : score = {}, N = {}'.format(
    result1[np.argmax(result1)], 
    np.argmax(result1) + 1))

```

```python
def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
```

```python
with open('q1_bestN', 'w') as f:
    f.write(str(np.argmax(result) + 1))
```

```python
with open('q2_score_bestN', 'w') as f:
    f.write(str(toFixed(result[np.argmax(result)], 2)))
```

```python
with open('q3_bestN_scale', 'w') as f:
    f.write(str(np.argmax(result1) + 1))
```

```python
with open('q4_score_bestN_scale', 'w') as f:
    f.write(str(toFixed(result1[np.argmax(result1)], 2)))
```
